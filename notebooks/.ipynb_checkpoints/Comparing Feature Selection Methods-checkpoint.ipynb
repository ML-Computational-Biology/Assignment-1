```python
# 0. Imports & Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.linear_model import ElasticNet, LassoCV
from sklearn.feature_selection import f_regression, mutual_info_regression, SelectKBest, RFE, SelectFromModel
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score
```

```python
# 1. Load scaled data
X_dev_df = pd.read_csv("../data/development_final_data.csv")
X_eval_df = pd.read_csv("../data/evaluation_final_data.csv")
```

```python
# 2. Separate features and target
y_dev = X_dev_df['BMI'].values
X_dev = X_dev_df.drop(columns=['BMI']).values
y_eval = X_eval_df['BMI'].values
X_eval = X_eval_df.drop(columns=['BMI']).values

# Save copy of feature names
feature_names = X_dev_df.drop(columns=['BMI']).columns
```

```python
# 3. Define evaluation function
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return {
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'MAE': mean_absolute_error(y_test, y_pred),
        'R2': r2_score(y_test, y_pred)
    }
```

```python
# 4. Define feature selection methods
feature_methods = {}

# LassoCV
lasso = LassoCV(cv=5, random_state=42).fit(X_dev, y_dev)
model_lasso = SelectFromModel(lasso, prefit=True)
feature_methods['LassoCV'] = model_lasso.get_support(indices=True)

# F-test
f_test = SelectKBest(score_func=f_regression, k=10).fit(X_dev, y_dev)
feature_methods['F-test'] = f_test.get_support(indices=True)

# Mutual Information
mi = SelectKBest(score_func=mutual_info_regression, k=10).fit(X_dev, y_dev)
feature_methods['Mutual Information'] = mi.get_support(indices=True)

# RFE with ElasticNet
rfe = RFE(estimator=ElasticNet(), n_features_to_select=10).fit(X_dev, y_dev)
feature_methods['RFE'] = rfe.get_support(indices=True)

# Random Forest Importance
rf = RandomForestRegressor(random_state=42).fit(X_dev, y_dev)
importance = rf.feature_importances_
rf_top_idx = np.argsort(importance)[-10:]
feature_methods['Random Forest'] = rf_top_idx
```

```python
# 5. Evaluate each method
results = {}

for method, idx in feature_methods.items():
    print(f"\nEvaluating method: {method}")
    X_train_sel = X_dev[:, idx]
    X_test_sel = X_eval[:, idx]

    model = ElasticNet()
    scores = evaluate_model(model, X_train_sel, y_dev, X_test_sel, y_eval)
    results[method] = scores
    print(f"RMSE: {scores['RMSE']:.4f}, MAE: {scores['MAE']:.4f}, R2: {scores['R2']:.4f}")
```

```python
# 6. Correlation Analysis Between Features and Target
X_features = X_dev_df.drop(columns=['BMI'])
y_target = X_dev_df['BMI']
correlations = X_features.corrwith(y_target)
top_corr = correlations.abs().sort_values(ascending=False).head(15)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_corr.values, y=top_corr.index)
plt.title("Top 15 Features by Absolute Pearson Correlation with BMI")
plt.xlabel("Absolute Correlation")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 7. Feature-to-Feature Correlation Heatmap
corr_matrix = X_features[top_corr.index].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", center=0)
plt.title("Correlation Matrix of Top Correlated Features")
plt.tight_layout()
plt.show()

print("\nFeature selection evaluation and correlation analysis completed.")
```
